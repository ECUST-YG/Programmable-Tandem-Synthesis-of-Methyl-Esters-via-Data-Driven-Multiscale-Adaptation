```python
import pandas as pd
import numpy as np
import re
```

```python
df1=pd.read_excel(r"Original_database.xlsx",sheet_name="Ti-Cs")
```

```python
pos_map = {
    1:(0,0), 2:(0,1), 3:(0,2), 4:(0,3),
    5:(1,1), 6:(1,2), 7:(1,3), 8:(1,4),
    9:(2,2), 10:(2,3), 11:(2,4), 12:(2,5),
    13:(3,3), 14:(3,4), 15:(3,5), 16:(3,6)
}

def catalyst_to_matrix(desc):
    if not isinstance(desc, str):
        return None
    mat = np.zeros((4, 7), dtype=int)
    
    if desc.strip() == '16Cs':
        for r, c in pos_map.values():
            mat[r, c] = 1
        return mat

    m = re.search(r'-(.+)', desc)
    if m is None:
        return None

    pos_str = m.group(1)
    ti_positions = list(map(int, pos_str.split('+')))

    for i in range(1, 17):
        r, c = pos_map[i]
        if i in ti_positions:
            mat[r, c] = 2 
        else:
            mat[r, c] = 1 
    return mat

df1['Matrix'] = df1['M1-M2'].apply(catalyst_to_matrix)
```

```python
import os
import ast
import random
import numpy as np
import pandas as pd

os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
os.environ['TF_DISABLE_MKL'] = '1'
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import tensorflow as tf
from sklearn.metrics import accuracy_score
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import ReduceLROnPlateau, LearningRateScheduler
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers

SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

BATCH_SIZE = 32
EPOCHS = 100
NUM_CLASSES = 4
EXPECTED_SHAPE = (4, 7)
L2 = 1e-4
DROPOUT = 0.3
INIT_LR = 1e-3

REDUCE_LR_CALLBACK = ReduceLROnPlateau(monitor='loss', factor=0.5, patience=8, min_lr=1e-6)

def create_adaptive_lr_schedule(initial_lr=INIT_LR):
    def lr_schedule(epoch):
        if epoch < 30:
            return initial_lr
        elif epoch < 60:
            return initial_lr * 0.3
        elif epoch < 80:
            return initial_lr * 0.2
        else:
            return initial_lr * 0.1
    return LearningRateScheduler(lr_schedule)

def parse_matrix_cell(cell):
    if isinstance(cell, np.ndarray):
        return cell
    if isinstance(cell, list):
        return np.array(cell)
    if isinstance(cell, (int, float)):
        return np.array([[cell]])
    if isinstance(cell, str):
        try:
            parsed = ast.literal_eval(cell)
            return np.array(parsed)
        except Exception:
            try:
                parts = [float(x) for x in cell.replace(';', ',').split(',') if x.strip() != '']
                return np.array(parts)
            except Exception:
                raise ValueError(f"Cannot parse Matrix cell: {cell[:80]}")
    raise ValueError(f"Unsupported Matrix cell type: {type(cell)}")

def dataset_to_Xy(dataset, expected_shape=EXPECTED_SHAPE):
    """Extract feature matrix and labels"""
    if 'Matrix' not in dataset.columns:
        raise KeyError("Dataset must contain 'Matrix' column")
    
    matrices = []
    for i, cell in enumerate(dataset['Matrix'].values):
        arr = parse_matrix_cell(cell)
        if arr.ndim == 1:
            if arr.size == expected_shape[0]*expected_shape[1]:
                arr = arr.reshape(expected_shape)
            else:
                raise ValueError(f"Sample {i} has incorrect size")
        elif arr.ndim == 2 and arr.shape != expected_shape:
            raise ValueError(f"Sample {i} has incorrect shape {arr.shape}")
        matrices.append(arr.astype(np.float32))
    
    X = np.stack(matrices, axis=0)[..., np.newaxis]
    
    y_raw = None
    if 'True Class' in dataset.columns:
        y_raw = dataset['True Class'].astype(int).values
    
    return X, y_raw

def build_model_improved(input_shape=(4,7), num_classes=NUM_CLASSES, dropout_rate=DROPOUT, l2_reg=L2, lr=INIT_LR):
    from tensorflow.keras.layers import Input, Conv2D, LeakyReLU, Concatenate, MaxPooling2D, GlobalAveragePooling2D, Dense, Dropout
    from tensorflow.keras.models import Model
    
    inp = Input(shape=(input_shape[0], input_shape[1], 1))
    x1 = Conv2D(96, (1,5), padding='same', kernel_regularizer=regularizers.l2(l2_reg))(inp)
    x1 = LeakyReLU(negative_slope=0.1)(x1)
    
    x2 = Conv2D(128, (1,5), padding='same', kernel_regularizer=regularizers.l2(l2_reg))(x1)
    x2 = LeakyReLU(negative_slope=0.1)(x2)
    x2 = Concatenate()([x1, x2])
    
    x3 = Conv2D(128, (1,5), padding='same', kernel_regularizer=regularizers.l2(l2_reg))(x2)
    x3 = LeakyReLU(negative_slope=0.1)(x3)
    x3 = Concatenate()([x1, x2, x3])
    
    x = MaxPooling2D((1,2))(x3)
    
    x = Conv2D(256, (2,2), padding='valid', kernel_regularizer=regularizers.l2(l2_reg))(x)
    x = LeakyReLU(negative_slope=0.1)(x)
    
    x = GlobalAveragePooling2D()(x)
    
    x = Dense(1024, kernel_regularizer=regularizers.l2(l2_reg))(x)
    x = LeakyReLU(negative_slope=0.1)(x)
    x = Dropout(dropout_rate)(x)
    
    x = Dense(512, kernel_regularizer=regularizers.l2(l2_reg))(x)
    x = LeakyReLU(negative_slope=0.1)(x)
    x = Dropout(dropout_rate)(x)
    
    out = Dense(num_classes, activation='softmax')(x)
    
    model = Model(inp, out)
    
    optimizer = Adam(learning_rate=lr, beta_1=0.9, beta_2=0.999, epsilon=1e-7)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model

def train_final_model(dataset, model_save_path="final_model.keras"):
    print("\n=== Training final model ===")
    print(f"Dataset shape: {dataset.shape}")
    
    X, y_raw = dataset_to_Xy(dataset)
    
    if y_raw is None:
        raise ValueError("Training data must contain 'True Class' column")
    
    y0 = (y_raw - 1).astype(int)
    y_cat = to_categorical(y0, num_classes=NUM_CLASSES)
    
    print(f"Training data shape: X{X.shape}, y{y_cat.shape}")
    print("Class distribution:", np.unique(y_raw, return_counts=True))
    
    model = build_model_improved()
    
    callbacks = [
        create_adaptive_lr_schedule(INIT_LR),
        REDUCE_LR_CALLBACK
    ]
    
    history = model.fit(
        X, y_cat,
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        callbacks=callbacks,
        verbose=1
    )
    
    model.save(model_save_path)
    print(f"Final model saved: {model_save_path}")
    
    train_predictions = model.predict(X, verbose=0)
    train_predicted_classes = np.argmax(train_predictions, axis=1) + 1
    train_accuracy = accuracy_score(y_raw, train_predicted_classes)
    print(f"Training accuracy: {train_accuracy:.4f}")
    
    return model, history

def predict_df_result(model, df_result, output_csv="df_result_predictions.csv"):
    print("\n=== Predicting new samples ===")
    print(f"df_result shape: {df_result.shape}")
    
    if 'Matrix' not in df_result.columns:
        raise KeyError("df_result must contain 'Matrix' column")
    
    X_new, _ = dataset_to_Xy(df_result)
    print(f"Feature shape: {X_new.shape}")
    
    predictions_proba = model.predict(X_new, verbose=1)
    predicted_classes = np.argmax(predictions_proba, axis=1) + 1
    
    df_result['Predicted_Class'] = predicted_classes
    df_result['Class_1_Probability'] = predictions_proba[:, 0]
    df_result['Class_2_Probability'] = predictions_proba[:, 1]
    df_result['Class_3_Probability'] = predictions_proba[:, 2]
    df_result['Class_4_Probability'] = predictions_proba[:, 3]
    df_result['Confidence'] = np.max(predictions_proba, axis=1)
    
    df_result.to_csv(output_csv, index=False)
    print(f"Prediction saved: {output_csv}")
    
    print("\n=== Prediction summary ===")
    print(f"Total samples: {len(df_result)}")
    
    class_counts = df_result['Predicted_Class'].value_counts().sort_index()
    for class_id, count in class_counts.items():
        print(f"Class {class_id}: {count} samples ({count/len(df_result)*100:.1f}%)")
    
    print(f"Mean confidence: {df_result['Confidence'].mean():.4f}")
    
    print("\n=== First 5 predictions ===")
    preview_cols = [
        'Predicted_Class', 'Class_1_Probability',
        'Class_2_Probability', 'Class_3_Probability',
        'Class_4_Probability', 'Confidence'
    ]
    print(df_result[preview_cols].head())
    
    return df_result

if __name__ == "__main__":
    print("=== CNN Matrix Classification - Final Training & Prediction ===")
    
    try:
        print("Checking data...")
        print(f"df type: {type(df)}, shape: {df.shape}")
        print(f"df_result type: {type(df_result)}, shape: {df_result.shape}")
        
        print("\n1. Training model...")
        trained_model, history = train_final_model(df)
        
        print("\n2. Predicting...")
        result_with_predictions = predict_df_result(
            model=trained_model,
            df_result=df_result,
            output_csv=r"Prediction_result.csv"
        )
        
        print("\nPreview:")
        print(result_with_predictions[['Predicted_Class', 'Confidence']].head())
        
    except NameError as e:
        print(f"Data error: {e}")
    except Exception as e:
        print(f"Execution error: {e}")
        import traceback
        traceback.print_exc()
    
    print("\nDone.")
```
