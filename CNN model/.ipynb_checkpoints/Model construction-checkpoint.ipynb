```python
df1=pd.read_excel(r"Original_database.xlsx",sheet_name="Ti-Cs")
```

```python
pos_map = {
    1:(0,0), 2:(0,1), 3:(0,2), 4:(0,3),
    5:(1,1), 6:(1,2), 7:(1,3), 8:(1,4),
    9:(2,2), 10:(2,3), 11:(2,4), 12:(2,5),
    13:(3,3), 14:(3,4), 15:(3,5), 16:(3,6)
}

def catalyst_to_matrix(desc):
    if not isinstance(desc, str):
        return None
    mat = np.zeros((4, 7), dtype=int)
    
    if desc.strip() == '16Cs':
        for r, c in pos_map.values():
            mat[r, c] = 1
        return mat

    m = re.search(r'-(.+)', desc)
    if m is None:
        return None

    pos_str = m.group(1)
    ti_positions = list(map(int, pos_str.split('+')))

    for i in range(1, 17):
        r, c = pos_map[i]
        if i in ti_positions:
            mat[r, c] = 2 
        else:
            mat[r, c] = 1 
    return mat

df1['Matrix'] = df1['M1-M2'].apply(catalyst_to_matrix)
```

```python
import numpy as np
import re

pos_map = {
    1:(0,0), 2:(0,1), 3:(0,2), 4:(0,3),
    5:(1,1), 6:(1,2), 7:(1,3), 8:(1,4),
    9:(2,2), 10:(2,3), 11:(2,4), 12:(2,5),
    13:(3,3), 14:(3,4), 15:(3,5), 16:(3,6)
}

def catalyst_to_matrix(desc):

    if not isinstance(desc, str):
        return None

    mat = np.zeros((4, 7), dtype=int)

    if desc.strip() == '16Ti':
        for r, c in pos_map.values():
            mat[r, c] = 2
        return mat

    m = re.search(r'-(.+)', desc)
    if m is None:
        return None

    pos_str = m.group(1)
    ti_positions = list(map(int, pos_str.split('+')))

    for i in range(1, 17):
        r, c = pos_map[i]
        if i in ti_positions:
            mat[r, c] = 2   
        else:
            mat[r, c] = 3   

    return mat


df3['Matrix'] = df3['M1-M2'].apply(catalyst_to_matrix)
```

```python
import os
import ast
import random
import numpy as np
import pandas as pd

os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
os.environ['TF_DISABLE_MKL'] = '1'
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import tensorflow as tf

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Conv2D, MaxPooling2D, Dense, Concatenate, Dropout,
    GlobalAveragePooling2D
)
from tensorflow.keras.callbacks import ReduceLROnPlateau, LearningRateScheduler
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers

SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

N_SPLITS = 5
BATCH_SIZE = 32
EPOCHS = 100
NUM_CLASSES = 4
RANDOM_STATE = SEED
EXPECTED_SHAPE = (4, 7)
L2 = 1e-4
DROPOUT = 0.3
INIT_LR = 1e-3

REDUCE_LR_CALLBACK = ReduceLROnPlateau(
    monitor='val_loss', factor=0.5, patience=8, min_lr=1e-6
)

def create_adaptive_lr_schedule(initial_lr=INIT_LR):
    def lr_schedule(epoch):
        if epoch < 30:
            return initial_lr
        elif epoch < 60:
            return initial_lr * 0.3
        elif epoch < 80:
            return initial_lr * 0.2
        else:
            return initial_lr * 0.1
    return LearningRateScheduler(lr_schedule)

def parse_matrix_cell(cell):
    if isinstance(cell, np.ndarray):
        return cell
    if isinstance(cell, list):
        return np.array(cell)
    if isinstance(cell, (int, float)):
        return np.array([[cell]])
    if isinstance(cell, str):
        try:
            return np.array(ast.literal_eval(cell))
        except:
            parts = [float(x) for x in cell.replace(';', ',').split(',') if x.strip() != '']
            return np.array(parts)
    raise ValueError(f"Unsupported Matrix cell type: {type(cell)}")

def dataset_to_Xy(dataset):
    X = np.stack([parse_matrix_cell(x).reshape(EXPECTED_SHAPE) for x in dataset['Matrix']], axis=0)[..., np.newaxis]
    y_raw = dataset['True Class'].astype(int).values
    return X, y_raw

def build_model_improved():
    from tensorflow.keras.layers import LeakyReLU

    inp = Input(shape=(4,7,1))

    x1 = Conv2D(96, (1,5), padding='same', kernel_regularizer=regularizers.l2(L2))(inp)
    x1 = LeakyReLU(0.1)(x1)

    x2 = Conv2D(128, (1,5), padding='same', kernel_regularizer=regularizers.l2(L2))(x1)
    x2 = LeakyReLU(0.1)(x2)
    x2 = Concatenate()([x1, x2])

    x3 = Conv2D(128, (1,5), padding='same', kernel_regularizer=regularizers.l2(L2))(x2)
    x3 = LeakyReLU(0.1)(x3)
    x3 = Concatenate()([x1, x2, x3])

    x = MaxPooling2D((1,2))(x3)

    x = Conv2D(256, (2,2), padding='valid', kernel_regularizer=regularizers.l2(L2))(x)
    x = LeakyReLU(0.1)(x)

    x = GlobalAveragePooling2D()(x)

    x = Dense(1024, kernel_regularizer=regularizers.l2(L2))(x)
    x = LeakyReLU(0.1)(x)
    x = Dropout(DROPOUT)(x)

    x = Dense(512, kernel_regularizer=regularizers.l2(L2))(x)
    x = LeakyReLU(0.1)(x)
    x = Dropout(DROPOUT)(x)

    out = Dense(NUM_CLASSES, activation='softmax')(x)

    model = Model(inp, out)
    model.compile(optimizer=Adam(INIT_LR), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def run_cv_and_plot(dataset):
    X, y_raw = dataset_to_Xy(dataset)
    y0 = (y_raw - 1).astype(int)
    y_cat = to_categorical(y0, NUM_CLASSES)

    skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=RANDOM_STATE)

    y_pred_full = np.zeros(len(dataset), dtype=int)
    y_score_full = np.zeros((len(dataset), NUM_CLASSES))

    accs = []

    for fold, (train_idx, test_idx) in enumerate(skf.split(X, y0), 1):
        print(f"Fold {fold}")

        model = build_model_improved()

        model.fit(
            X[train_idx], y_cat[train_idx],
            validation_data=(X[test_idx], y_cat[test_idx]),
            epochs=EPOCHS,
            batch_size=BATCH_SIZE,
            callbacks=[create_adaptive_lr_schedule(), REDUCE_LR_CALLBACK],
            verbose=0
        )

        preds = model.predict(X[test_idx], verbose=0)
        labels = np.argmax(preds, axis=1)

        y_pred_full[test_idx] = labels
        y_score_full[test_idx] = preds

        acc = accuracy_score(np.argmax(y_cat[test_idx], axis=1), labels)
        accs.append(acc)
        print(f"Fold accuracy: {acc}")

    print("Per-fold accuracy:", accs)

    overall_acc = accuracy_score(y0, y_pred_full)
    print(f"Overall accuracy: {overall_acc}")

    oof_df = pd.DataFrame({
        'True_Class': y_raw,
        'Predicted_Class': y_pred_full + 1
    })

    for i in range(NUM_CLASSES):
        oof_df[f'prob_class_{i+1}'] = y_score_full[:, i]

    oof_df['Matrix'] = dataset['Matrix'].astype(str)

    oof_df.to_csv("CV result.csv", index=False)
    print("Saved as CV result.csv")

if __name__ == "__main__":
    run_cv_and_plot(df)
```
